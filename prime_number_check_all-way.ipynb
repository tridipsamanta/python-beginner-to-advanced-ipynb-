## 1️⃣ Prime Check Using Simple Loop (Basic)

```python
num = int(input("Enter a positive number : "))
if num <= 1:
    print("Not Prime..")
else:
    for i in range(2,num):
        if num%i==0:
            print("Not Prime..")
            break
    else:
        print("Prime...")
```

## 2️⃣ Prime Check Using Flag Variable

```python
num = int(input("Enter a positive number : "))
flag = True
if num <= 1:
    flag = False
else:
    for i in range(2,num):
        if num%i==0:
            flag = False
            break
    
if flag:
    print('Prime')
else:
    print('not prime')
```

## 3️⃣ Prime Check Using while Loop

```python
num = int(input("Enter a positive number : "))
i=2
if num <= 1:
    print("Not Prime..")
else:
    while i < num:
        if num%i==0:
            print("Not Prime..")
            break
        i += 1
    else:
        print("Prime...")
```

## 4️⃣ Optimized Prime Check (Up to √n) ⭐

```python
import math
num = int(input("Enter a positive number : "))
if num <= 1:
    print("Not Prime..")
else:
    for i in range(2,int(math.sqrt(num))+1):
        if num%i==0:
            print("Not Prime..")
            break
        i += 1
    else:
        print("Prime...")
```

## 5️⃣ Prime Check Using Function

```python
import math
def prime_check(num):
    if num <= 1:
        return False
    else:
        for i in range(2,int(math.sqrt(num))+1):
            if num%i==0:
                return False
                break
        else:
            return True
num = int(input("Enter a positive number : "))
result = prime_check(num)

print(result)

```

## 6️⃣ Prime Check Using Recursion

```python
def is_prime(n,i=2):
    if n <= 1:
        return False
    if i*i > n:
        return True
    if n % i == 0:
        return False
    return is_prime(n,i+1)

print(is_prime(17))
```
