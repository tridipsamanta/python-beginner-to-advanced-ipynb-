## HCF Using Loop (Basic)

```python
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))

hcf = 0

for i in range(1,min(a,b)+1):
    if a%i==0 and b%i==0:
        hcf = i
print("HCF : ",hcf)
```

## 2️⃣ HCF Using Euclidean Algorithm (Efficient ⭐) 

```python
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))

while b!= 0:
    a,b = b,a%b

print("HCF : ",a)
```

## 3️⃣ HCF Using Function

```python
def hcf(a,b):
    while b!= 0:
        a,b = b,a%b
    return a
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))
print("HCF : ",hcf(a,b))
```

## 4️⃣ HCF Using math.gcd() (Built-in)

```python
import math

a = int(input("Enter first number : "))
b = int(input("Enter second number : "))

print("HCF : ",math.gcd(a,b))
```

# LCM Programs

## 5️⃣ LCM Using Loop

```python
a = int(input("Enter first number : "))
b = int(input("Enter second number : "))

lcm = max(a,b)
while True:
    if lcm%a==0 and lcm%b==0:
        print("LCM : ",lcm)
        break
    lcm += 1
```

## 6️⃣ LCM Using Formula (Best Method ⭐)

```python
import math

a = 12
b = 18

lcm = (a * b) // math.gcd(a, b)
print("LCM:", lcm)
```

## 7️⃣ LCM Using Function

```python
def lcm(a, b):
    return (a * b) // hcf(a, b)

print(lcm(15, 20))
```

## 8️⃣ LCM Using math.lcm() (Python 3.9+)

```python
import math
a = 12
b = 18

print("LCM : ",math.lcm(a,b))
```
