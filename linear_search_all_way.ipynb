## 1️⃣ Basic Linear Search

```python
n = int(input("Enter array length : "))
arr = []
print("Enter elements : ")
for i in range(n):
    arr.append(int(input()))
key = int(input("Enter key element : "))
for i in range(n):
    if arr[i] == key:
        print("Elements found at index : ",i)
        break
else:
    print("Elemnt not found...")
```

## 2️⃣ Linear Search Using Function

```python
def linear_search(arr,key):
    for i in range(len(arr)):
        if arr[i]==key:
            return i
            break
    else:
        return -1
n = int(input("Enter array length : "))
arr = []
print("Enter elements : ")
for i in range(n):
    arr.append(int(input()))
key = int(input("Enter key element : "))
if not -1:
    print("Element found at index : ",linear_search(arr,key))
else:
    print("Element not found")
```

## 3️⃣ Linear Search Using While Loop

```python
n = int(input("Enter array length : "))
arr = []
print("Enter elements : ")
for i in range(n):
    arr.append(int(input()))
key = int(input("Enter key element : "))

i = 0
while i <= n:
    if arr[i] == key:
        print("Element found at index : ",i)
        break
    i += 1
else:
    print("Element not found...")
```
