```python
import matplotlib.pyplot as plt


```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# Set background color
ax.set_facecolor('#013243')  # Dark blue background
fig.patch.set_facecolor('#013243')

# Add text for NumPy logo
ax.text(0.5, 0.5, 'NumPy', color='white', fontsize=60,
        fontweight='bold', ha='center', va='center', family='sans-serif')

# Remove axes
ax.axis('off')

# Show the plot
plt.show()
```
