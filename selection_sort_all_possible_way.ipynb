## 1️⃣ Basic Selection Sort (Ascending Order)

```python
arr = [4,5,6,1,8,9]
n = len(arr)

for i in range(n):
    min_index = i
    for j in range(i+1,n):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i],arr[min_index] = arr[min_index],arr[i]

print(arr)
```

## 2️⃣ Selection Sort (Descending Order)

```python
arr = [4,5,1,3,7]
n = len(arr)
for i in range(n):
    min_index = i
    for j in range(i+1,n):
        if arr[j] > arr[min_index]:
            min_index = j
    arr[i],arr[min_index] = arr[min_index],arr[i]
print(arr)
```

## 3️⃣ Selection Sort Using Function

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1,n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i],arr[min_index] = arr[min_index],arr[i]
    return arr
arr = [3,5,6,7,1,2]
print(selection_sort(arr))
```

## 5️⃣ Selection Sort Using While Loop

```python
arr = [5, 3, 1, 4, 2]
n = len(arr)
i = 0

while i < n:
    min_index = i
    j = i + 1
    while j < n:
        if arr[j] < arr[min_index]:
            min_index = j
        j += 1
    arr[i], arr[min_index] = arr[min_index], arr[i]
    i += 1

print(arr)
```

## Recursive Selection Sort (Ascending Order)

```python
def selection_sort_recursive(arr, start, n):
    # Base case: when starting index reaches end
    if start >= n - 1:
        return

    # Find index of minimum element
    min_index = start
    for i in range(start + 1, n):
        if arr[i] < arr[min_index]:
            min_index = i

    # Swap minimum with first element of unsorted part
    arr[start], arr[min_index] = arr[min_index], arr[start]

    # Recursive call for remaining array
    selection_sort_recursive(arr, start + 1, n)


# Driver code
arr = [64, 25, 12, 22, 11]
selection_sort_recursive(arr, 0, len(arr))
print(arr)
```
