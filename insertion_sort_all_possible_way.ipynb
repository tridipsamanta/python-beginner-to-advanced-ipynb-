## 1️⃣ Basic Insertion Sort (Ascending Order)

```python
arr = [5,3,2,4,1]
n = len(arr)

for i in range(1,n):
    key = arr[i]
    j = i-1

    while j>=0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1
    arr[j+1] = key
print(arr)
```

## 2️⃣ Insertion Sort (Descending Order)

```python
arr = [1,2,3,4,5,6]
n = len(arr)

for i in range(1,n):
    key = arr[i]
    j = i-1

    while j >= 0 and arr[j] < key:
        arr[j+1] = arr[j]
        j -= 1
    arr[j+1] = key
print('after sorting : ',arr)
```

## 3️⃣ Optimized Insertion Sort (Early Stop)

```python
arr = [1,2,3,4,5,6]
n = len(arr)
# nsertion sort is already optimized by nature, 
# but this version minimizes unnecessary checks.
for i in range(1,n):
    key = arr[i]
    j = i-1
    if arr[j] > key: # new line code 
     while j >= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1
     arr[j+1] = key
print('after sorting : ',arr)
```

## 4️⃣ Insertion Sort Using Function

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1,n):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
arr = [1,2,3,4,5]
print(arr)
```

## 5️⃣ Insertion Sort Using While Loop Only

```python
arr = [7, 4, 2, 9, 1,3]
i = 1

while i < len(arr):
    key = arr[i]
    j = i - 1

    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1

    arr[j + 1] = key
    i += 1

print(arr)
```

## 6️⃣ Insertion Sort for User Input

```python
n = int(input("Enter array length : "))
arr = []
print("Enter elements : ")
for i in range(n):
    arr.append(int(input()))
for i in range(1,n):
    key = arr[i]
    j = i-1

    while j>=0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -= 1
    arr[j+1] = key
print(arr)
```

## 7️⃣ Insertion Sort recursive .....

```python
def insertion_sort_recursive(arr,n):
    if n<=1:
        return
    
    insertion_sort_recursive(arr,n-1)

    key = arr[n-1]
    j = n-2

    while j>= 0 and arr[j] > key:
        arr[j+1] = arr[j]
        j -=1
    arr[j+1] = key


n = int(input("Enter array length : "))
arr = []
print("Enter elements : ")
for i in range(n):
    arr.append(int(input()))

insertion_sort_recursive(arr,n)
print(arr)
```
