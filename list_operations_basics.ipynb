## Write a program to find all duplicate elements in a list.

```python
lst = [1, 2, 3, 2, 4, 1, 5]
duplicates = []

for i in lst:
    if lst.count(i) > 1 and i not in duplicates:
        duplicates.append(i)

print("Duplicates:", duplicates)
```

## Write a program to find the Longest Increasing Subsequence (LIS).

```python
lst = [10, 22, 9, 33, 21, 50]
n = len(lst)
dp = [1] * n

for i in range(n):
    for j in range(i):
        if lst[i] > lst[j]:
            dp[i] = max(dp[i], dp[j] + 1)

print("Length of LIS:", max(dp))
```

## Write a program to sort a list without using sort().

```python
lst = [5, 2, 9, 1, 3]

for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] > lst[j]:
            lst[i], lst[j] = lst[j], lst[i]

print(lst)
```
