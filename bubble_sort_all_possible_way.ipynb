## Bubble Sort....

```python
# Basic Bubble sort ascending order..
arr = [10,30,50,2,7,8,9]
n = len(arr)

for i in range(n):
    for j in range(0,n-i-1):
        if arr[j] > arr[j+1]:
           temp = arr[j]
           arr[j] = arr[j+1]
           arr[j+1] = temp
print(arr)
```

## Bubble Sort descending order...

```python
# Basic Bubble sort descending order..
arr = [10,30,50,2,7,8,9]
n = len(arr)

for i in range(n):
    for j in range(0,n-i-1):
        if arr[j] < arr[j+1]:
           temp = arr[j]
           arr[j] = arr[j+1]
           arr[j+1] = temp
print(arr)
```

## Optimized Bubble Sort (If no swapping happens, the list is already sorted â†’ stop early.)

```python
# Basic Bubble sort descending order..
arr = [1,2,3,4,5,6]
n = len(arr)

for i in range(n):
    swap = False
    for j in range(0,n-i-1):
        if arr[j] > arr[j+1]:
           temp = arr[j]
           arr[j] = arr[j+1]
           arr[j+1] = temp
           swap = True
    if not swap:
        break    
print(arr)
```

## Bubble sort using function......

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j] > arr[j+1]:
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
    print(arr)

print(bubble_sort([90,1,5,100,600,300,7,9]))
```

## Bubble Sort using while loop...

```python
# Basic Bubble sort descending order..
arr = [10,30,50,2,7,8,9]
n = len(arr)
i=0
while i < n:
    j = 0
    while j < n-i-1:
        if arr[j] > arr[j+1]:
            temp = arr[j]
            arr[j] = arr[j+1]
            arr[j+1] = temp
        j += 1
    i += 1
print(arr)
```

```python
def bubble_sort_recusrive(arr,n):
    if n == 1:
        return
    
    for i in range(n-1):
        if arr[i] > arr[i+1]:
            arr[i] = arr[i+1]
            arr[i+1] = arr[i]
    
    bubble_sort_recursive(arr,n-1)

# n = int(input("Enter array length : "))
# arr = []
# for i in range(n):
#     arr.append(int(input()))
arr  = [1,2,7,3,6,9]
bubble_sort_recursive(arr,6)
print("After sorting : ",arr)
```

## Optimized Recursive Bubble Sort (Early Exit)

```python
def bubble_sort_recusrive(arr,n):
    if n == 1:
        return
    
    swapped = False

    for i in range(n-1):
        if arr[i] > arr[i+1]:
            arr[i] = arr[i+1]
            arr[i+1] = arr[i]
            swapped = True
    
    bubble_sort_recursive(arr,n-1)

    if not swapped:
        return

n = int(input("Enter array length : "))
arr = []
for i in range(n):
    arr.append(int(input()))
bubble_sort_recursive(arr,n)
print("After sorting : ",arr)
```
