## Create a Dictionary and print.

```python
student = {
    'name':'Tridip samanta',
    'Age':20,
    'Dept':'BCA'
}
print('Name : ',student['name'],'\nDepartment : ',student['Dept'])
```

## Access Dictionary values...

```python
student = {
    'name':'Tridip samanta',
    'Age':20,
    'Dept':'BCA'
}
print ('My age is : ',student['Age'])
```

## Add and update dictionary values..

```python
student = {
    'name':'Tridip samanta',
    'Age':20,
    'Dept':'BCA'
}
student['Age'] = 19
student['City'] = 'Kolkata'
print('My city is : ',student['City'])
```

## Remove ellements from dictionary...

```python
student = {
    'name':'Tridip samanta',
    'Age':20,
    'Dept':'BCA',
    'City':'Kolkata'
}
student.pop('City')
# or ............
del student['Dept']
print(student)
```

## Iterate through Dictionary...

```python
student = {
    'name':'Tridip samanta',
    'Age':20,
    'Dept':'BCA',
    'City':'Kolkata'
}
for key,value in student.items():
    print ( key ,' : ',value)
```

## Count Frequency of Elements (Important ‚≠ê)


```python
text = 'Banana'
freq = {}
for i in text:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
print(freq)
```

## Find Student with Highest Marks

```python
students = {
    'Tridip':80,
    'Sourav':90,
    'Somnath':90,
    'Gourav':80
}
topper = max(students,key=students.get)
print('Topper : ',topper)
```

## Merge Two Dictionaries

```python
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}

d3 = d1 | d2   # Python 3.9+
print(d3)
```

## Dictionary Comprehension

```python
squar = { x : x*x for x in range(1,10)}
print(squar)
```
